```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)  #sinwave
z=np.cos(x)  #cos wave
plt.plot(x,y,'gx')
plt.plot(x,z,'r--')
plt.plot(y,z)
# Adding Labels For the Axis
plt.xlabel("Angle")
plt.ylabel("Sin Value")
plt.title("Sinewave And Cos Wave")
plt.show()
```

```python
#Ploting THe Parabola
x=np.linspace(-100,100,20)
y=x**2
plt.plot(x,y)
plt.show
```

```python
#bar ploting
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
lang=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(lang,people)
plt.xlabel('LANGUAGE')
plt.ylabel('PEOPLE')
plt.show
```

```python
# Pie Chart
fig1=plt.figure()
ay=fig1.add_axes([0,0,1,1])
lang=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ay.pie(people,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(y)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='y')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D  Scatter Plot
fig3=plt.figure()
te=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.sin(z)
ax.scatter(x,y,z,cmap='Blue')
plt.show
```
